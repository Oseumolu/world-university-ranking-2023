```python
df['International Student'] = df['International Student'].str.rstrip('%')
```

```python
df['International Student'] = pd.to_numeric(df['International Student'], errors='coerce')
df.dropna(subset=['International Student'], inplace=True)
```

```python
df['International Student'] = pd.to_numeric(df['International Student'], errors='coerce')
df['International Student'].fillna(0, inplace=True)
```

```python
mean_value = df['International Student'].mean()
median_value = df['International Student'].median()
std_deviation = df['International Student'].std()
min_value = df['International Student'].min()
max_value = df['International Student'].max()

print("Mean:", mean_value)
print("Median:", median_value)
print("Standard Deviation:", std_deviation)
print("Min Value:", min_value)
print("Max Value:", max_value)
```

```python
import matplotlib.pyplot as plt

# Histogram
plt.hist(df['International Student'], bins=20, edgecolor='k')
plt.xlabel('International Student Percentage')
plt.ylabel('Frequency')
plt.title('Distribution of International Student Percentage')
plt.show()

# Box plot
plt.boxplot(df['International Student'])
plt.ylabel('International Student Percentage')
plt.title('Box Plot of International Student Percentage')
plt.show()

# Scatter plot (if comparing with another variable)
plt.scatter(df['Industry Income Score'], df['International Student'])
plt.xlabel('Industry Income Score')
plt.ylabel('International Student Percentage')
plt.title('Scatter Plot of International Student Percentage vs. Some Other Variable')
plt.show()
```
